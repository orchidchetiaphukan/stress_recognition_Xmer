# Loading the Data

```python
x_train  = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/x_train.npy")
x_test = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/x_test.npy")
y_train = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/y_train.npy")
y_test  = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/y_test.npy")
x_whole = np.concatenate((x_train , x_test),axis = 0 )
np.save("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/x_whole.npy", x_whole)
print(x_whole.shape)
y_whole  = np.concatenate((y_train , y_test), axis = 0)
np.save("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/y_whole.npy", y_whole)
print(y_whole.shape)
```

```python
x = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/x_whole.npy")
print(x.shape)
y = np.load("/content/drive/MyDrive/Ghanapriya Ma'am/WESAD/y_whole.npy")
y = pd.DataFrame(y)
print(y.shape)
```

```python
print("Before Replacing the labels:\n",pd.DataFrame(y).value_counts())
y_new = y.replace([1,2,3], [0,1,2])
print("After Repalacing the labels:\n", y_new.value_counts())
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split( x, y_new, test_size=0.1, random_state=10)
print(features_train.shape)
print(target_train.shape)
print(features_test.shape)
print(target_test.shape)
print()
print(pd.DataFrame(target_train).value_counts())
print(target_train[0:5])
print()
print(pd.DataFrame(target_test).value_counts())
print(target_test[0:5])
```

```python
# Training set class ratio for weight:
pd.DataFrame(target_train).value_counts()
```

```python
target_trainoe = tensorflow.keras.utils.to_categorical(target_train)
print(target_trainoe[0:5])
print()
target_testoe = tensorflow.keras.utils.to_categorical(target_test)
print(target_testoe[0:5])
```
